```python
import datetime
import pandas as pd

today_date=datetime.date.today()
if today_date.month<10:
    mes="0"+str(today_date.month)
else:
    mes=+str(today_date.month)
if today_date.day<10:
    dia="0"+str(today_date.day)
else:
    dia=str(today_date.day)
df=pd.read_excel(f"https://www.xm.com.co/pubregistros/PubFC{today_date.year}-"+mes+"-"+dia+".xlsx")
operadores_de_red=["CENTRALES ELECTRICAS DE NARIÑO S.A. E.S.P. - DISTRIBUIDOR"]
df=df[df["Operador de Red"].isin(operadores_de_red)]
filename="Registros_"+today_date.strftime("%d_%m_%Y")+".xlsx"
df.to_excel("Registros/"+filename,index=False)
```

```python
def resumen_reg(month,year):
    day=1
    prov_df=pd.read_excel("Resumenes/Example.xlsx")
    if month<10:
        month="0"+str(month)
    else:
        month=str(month)
    while True:
        if day<10:
            day_s="0"+str(day)
        else:
            day_s=str(day)
        try:
            df=pd.read_excel(f"https://www.xm.com.co/pubregistros/PubFC{year}-"+month+"-"+day_s+".xlsx")
            operadores_de_red=["CENTRALES ELECTRICAS DE NARIÑO S.A. E.S.P. - DISTRIBUIDOR"]
            df=df[df["Operador de Red"].isin(operadores_de_red)]
            prov_df=prov_df.append(df,ignore_index=True)
            day=day+1
        except:
            break
    return prov_df.drop_duplicates(subset=["Código SIC","Tipo de Solicitud","Estado"])
resumen_reg(3,2021)
```
